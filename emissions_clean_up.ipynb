```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files
emissions_data = "data/owid-co2-data (1).csv"
lat_lng_data = "data/countries_codes_and_coordinates.csv"



# Read the mouse data and the study results
emission_data = pd.read_csv(emissions_data)
coordinates_data = pd.read_csv(lat_lng_data)


#demographics
emission_data.head()
```

```python
coordinates_data.drop (["Alpha-2 code", "Numeric code"], axis=1, inplace=True)
coordinates_data.head()
```

```python
coordinates_data = coordinates_data.applymap(lambda x: x.replace('"', ''))
coordinates_data.head()
```

```python
coordinates_data = coordinates_data.rename(
    columns={
        "Country": 'country',
        "Alpha-3 code": "iso_code",
    })
coordinates_data.head()
```

```python
iso_code_data = coordinates_data['iso_code'].str.strip()
coordinates_data.head()
```

```python
coordinates_data['iso_code'] = iso_code_data
```
